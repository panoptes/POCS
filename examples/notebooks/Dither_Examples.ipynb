```python
import astropy.units as u
from astropy.coordinates import SkyCoord
from IPython.display import Image
```

```python
from pocs.utils import dither
```

# Create a base position

Chose an arbitrary sky position close to the South Celestial Pole to demonstrate correct handing of angular separations and projections.

```python
base = SkyCoord('4h23m45.3s -87d23m56s')
```

```python
base
```

```python
base.to_string('hmsdms')
```

# Dice 5 pattern dither, no random offsets

```python
dice5_positions = dither.get_dither_positions(base_position=base, 
                                              n_positions=5, 
                                              pattern=dither.dice5,
                                              pattern_offset=30 * u.arcminute,
                                              plot='dice5.png')
```

```python
dice5_positions
```

```python
dice5_positions.to_string('hmsdms')
```

```python
Image('dice5.png')
```

# Random offsets only

```python
random_positions = dither.get_dither_positions(base_position=base, 
                                               n_positions=20, 
                                               random_offset=0.5 * u.degree, 
                                               plot='random.png')
```

```python
random_positions
```

```python
random_positions.to_string('hmsdms')
```

```python
Image('random.png')
```

# Dice 9 pattern dither with random offsets

```python
dice9_positions = dither.get_dither_positions(base_position=base, 
                                              n_positions=20, 
                                              pattern=dither.dice9, 
                                              pattern_offset=30*u.arcminute, 
                                              random_offset=100*u.arcsecond, 
                                              plot='dice9.png')
```

```python
dice9_positions
```

```python
dice9_positions.to_string('hmsdms')
```

```python
Image('dice9.png')
```
